```python
import pandas as pd
```

```python
def getVotingResult(df):

    optionList = ['Adaptation WITHIN Ordinary', 'Chances WITHIN Reach/Doorstep', 'Data WITHIN Users', 
                 'Decentralization WITHIN Resilience', 'Flexibility WITHIN Resilience', 'Individual WITHIN Crowd',
                 'Normality WITHIN Resilience', 'Resilience WITHIN Order', 'Space WITHIN Epidemic', 'Parking WITHIN Reach']

    # list all voters and options
    votees = list(df)
    del votees[0]
    votees

    # list all options
    options = []
    for i in range(10):
        options.append(str(i+1))

    #list all voters
    voters = [ 'A','B','C','D','E','F','G','H','I','J','K','L']

    df = df.drop(columns = ['name'])
    df['total'] = df.apply(lambda x: x.sum(), axis=1)  # count the total ballots of every voter

    for i in range(len(df)):
        df.iloc[i] = df.iloc[i]/df['total'][i]*100  # normalize the total ballots of every voter to 100

    # df1 for options, df2 for voters    
    df1 = df.iloc[:, 0:10]
    df2 = df.iloc[:,10:-1]

    df1['counts'] = df1.apply(lambda x: x.sum(), axis=1)  # calculate how many ballots every voter give to options
    df2.loc['ballots'] = df2.apply(lambda x: x.sum()) # calculate how many ballots every voter get from voters

    # calculate how many ballots every voter give to options after being weighted by the ballots from voters
    df1['realCounts'] = df1['counts']
    for i in range(len(df)):
        df1['realCounts'][i] += df2[voters[i]]['ballots']

    # calculate how many ballots every option per voter gets after being weighted by the ballots from voters    
    for i in range(len(df1)):
        for option in options:
            df1[option][i] = df1[option][i]*df1['realCounts'][i]/df1['counts'][i]

    # calculate final total ballots of every option 
    result = {}
    for option in options:
        result[option] = sum(df1[option])

    results = {}
    for i in range(len(options)):
        results[optionList[i]] = sum(df1[options[i]])

    # get the best 2 options
    btos = sorted(list(result.values()), reverse = True)
    first = list(result.values()).index(btos[0])
    second = list(result.values()).index(btos[1])
    print('The first option is: '+ optionList[first])
    print('The second option is: '+ optionList[second])
    print('===================================================')
    print('The complete results:')
    return results
```

```python
df = pd.read_csv('./data.csv')
df
```

```python
getVotingResult(df)
```
